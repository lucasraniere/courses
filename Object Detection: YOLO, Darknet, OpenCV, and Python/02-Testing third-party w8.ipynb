```python
import cv2
import matplotlib.pyplot as plt

def mostrar(img_path):
    img = cv2.imread(img_path)
    img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)
    fig = plt.gcf()
    fig.set_size_inches(18,10)
    plt.axis("off")
    plt.imshow(img)
    plt.show()


mostrar("predictions.jpg")
```
